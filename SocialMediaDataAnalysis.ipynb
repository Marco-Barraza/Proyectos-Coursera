```python
#import librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sb
```

```python
#cargar el dataset
df = pd.read_csv("/home/jovyan/work/WhatsgoodlyData-6.csv")
```

```python
#Se buscan valores nulos
df.isnull().sum()
```

```python
#Exploraci√≥n de columnas
df.head()
```

```python
#Analisar los tipos de datos
df.info()
```

```python
cat = df.groupby("Answer")[["Percentage"]].mean().reset_index()
sb.set(style="whitegrid")
ax = sb.barplot(x="Answer", y="Percentage", data=cat)
"En promedio la red social m√°s usada no est√° identificada."
```

```python
cat = df.groupby("Segment Type")[["Count"]].mean().reset_index()
sb.set(style="whitegrid")
ax = sb.barplot(x="Segment Type", y="Count", data=cat)
"El segmento m√°s frecuente son los usuarios conectados a trav√©s del m√≥vil."
```

```python
ax = sb.boxplot(x="Segment Type", y="Count", data=df)
"La dispersi√≥n en los usuarios conectados a trav√©s del m√≥vil es significativamente normal"
```

```python
#Filtramos por el subgrupo m√≥vil
df.drop(df[df["Segment Type"] != "Mobile"].index, inplace=True)
```

```python
# Combinar dos grupos de la variable 'Answer'
df.loc[df['Answer'].isin(['Facebook', "Instagram"]), 'Answer'] = 'Fac/Int'

# Calcular la suma de los valores para la nueva categor√≠a
df_nueva = df.groupby('Answer', as_index=False).sum()
```

```python
# Crear el gr√°fico circular
plt.figure(figsize=(8, 8))
plt.pie(df["Count"], labels=df["Answer"], autopct='%1.1f%%', startangle=140)

# A√±adir t√≠tulo
plt.title('Gr√°fico Circular')

# Mostrar el gr√°fico
plt.axis('equal')  # Asegurar que el gr√°fico sea circular
plt.show()
```

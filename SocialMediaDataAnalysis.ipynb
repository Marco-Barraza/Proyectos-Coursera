Pandas is a library used for data manipulation and analysis. Numpy is a library used for numerical computations. Matplotlib is a library used for data visualization. Seaborn is a library used for statistical data visualization. Random is a library used to generate random numbers.

```python
#import librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sb
```

```python
#cargar el dataset
df = pd.read_csv("/home/jovyan/work/WhatsgoodlyData-6.csv")
```

```python
#Se buscan valores nulos
df.isnull().sum()
```

```python
#Exploración de columnas
df.head()
```

```python
#Analisar los tipos de datos
df.info()
```

```python
cat = df.groupby("Answer")[["Percentage"]].mean().reset_index()
sb.set(style="whitegrid")
ax = sb.barplot(x="Answer", y="Percentage", data=cat)
"En promedio la red social más usada no está identificada."
```

```python
cat = df.groupby("Segment Type")[["Count"]].mean().reset_index()
sb.set(style="whitegrid")
ax = sb.barplot(x="Segment Type", y="Count", data=cat)
"El segmento más frecuente son los usuarios conectados a través del móvil."
```

```python
ax = sb.boxplot(x="Segment Type", y="Count", data=df)
"La dispersión en los usuarios conectados a través del móvil es significativamente normal"
```

```python
#Filtramos por el subgrupo móvil
df.drop(df[df["Segment Type"] != "Mobile"].index, inplace=True)
```

```python
# Combinar dos grupos de la variable 'Answer'
df.loc[df['Answer'].isin(['Facebook', "Instagram"]), 'Answer'] = 'Fac/Int'

# Calcular la suma de los valores para la nueva categoría
df_nueva = df.groupby('Answer', as_index=False).sum()
```

```python
# Crear el gráfico circular
plt.figure(figsize=(8, 8))
plt.pie(df["Count"], labels=df["Answer"], autopct='%1.1f%%', startangle=140)

# Añadir título
plt.title('Gráfico Circular')

# Mostrar el gráfico
plt.axis('equal')  # Asegurar que el gráfico sea circular
plt.show()
```
